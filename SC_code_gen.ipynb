```python
from collections import deque
import random

def zerogen(n1,n2): #генератор нулевой матрицы
    blockline=[0 for i0 in range(n2)]
    return [blockline for i1 in range(n1)]

def cyclegen(size,baseline,rotatenum=1): #генератор циклических матриц из 1 строчки
    matrix=[]
    matrix.append(baseline)
    newline=deque(baseline)
    for i in range(size-1):
        newline.rotate(rotatenum)
        matrix.append(list(newline))
    return matrix

def Hgen(H): #Функция генерации Hi
    Hres=[]   
    block=[]
    blockline=[]
    for i in range(y):
        base=deque(protoline)
        for j in range(k):
            if H[i][j]==0:
                block=zerogen(k,k)
            else:
                base=deque(protoline)
                base.rotate(-L[i][j])
                block.append(list(base))
                for i2 in range(1,k):
                    base=deque(base)
                    base.rotate(1)
                    block.append(list(base))
            Hres.append(block)
            block=[]
            blockline=[]
    return Hres

def recombinatorH(H): #Функция для вывода, соединяет массивы Hi в 1 матрицу
    single=[]
    newM=[]
    for i in range(len(H[0])):
        for j in range(len(H)/2):
            single=single+H[j][i]
        newM.append(single)
        single=[]
    for i in range(len(H[0])):
        for j in range(len(H)/2,len(H)):
            single=single+H[j][i]
        newM.append(single)
        single=[]
    return newM

def recombinatorHpar(H): #Функция для вывода, соединяет массивы H в 1 матрицу
    single=[]
    newM=[]
    for i in range(len(H[0])):
        for j in range(len(H)):
            single=single+H[j][i]
        newM.append(single)
        single=[]
    return newM

def HPigen(i): #Генератор матрицы Hpi
    line=[]
    matrix=[]
    for s in range(y):
        for t in range(k):
            if P[s][t]==i:
                el=B[s][t]
            else:
                el=0
            line.append(el)
        matrix.append(line)
        line=[]
    return matrix

def Lgen(z1): #генерация случайной матрицы L, можно также генерировать и B и P
    line=[]
    matrix=[]
    for s in range(y):
        for t in range(k):
            el=randint(0,z1-1)
            line.append(el)
        matrix.append(line)
        line=[]
    return matrix    

z=3            
y=2
k=3
m=1
L0=4
protoline=[1,0,0] #Первая линия протографа сигма
#Задание матриц
B=[[1,1,1],[1,1,1]]
P=[[0,1,0],[1,0,1]]
L=[[1,0,2],[0,2,1]]
Hp0=HPigen(0)
Hp1=HPigen(1)

Zero1 = zerogen(2,3)
Zero2 = zerogen(6,9)
H0=Hgen(Hp0)
H1=Hgen(Hp1)
```
